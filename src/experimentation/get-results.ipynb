```python
from azureml.core import Experiment
from azureml.core import Workspace
import pandas as pd
from azureml.core import Run

ws = Workspace(subscription_id='3ad55f04-c0b2-4357-aea6-5b100ab15e50', resource_group='azuredscert', workspace_name='az-ml-dscert-ws')
experiment = Experiment(workspace=ws, name='train-diabetes-classification')
runs = experiment.get_runs()

run_ids = [run.id for run in runs]

# Initialize an empty list to store the metrics for each run
all_metrics = []

# Iterate over the runs
for run_id in run_ids:
    run = Run(experiment, run_id)
    # Check if the run was successful
    if run.get_status() == 'Completed':
        # Get run metrics
        metrics = run.get_metrics()

        # Get reg_rate from the run's tags
        reg_rate = run.get_tags().get('reg_rate')

        # Add the reg_rate parameter to the metrics
        if reg_rate is not None:
            metrics['reg_rate'] = float(reg_rate)

        # Add the metrics to the list
        all_metrics.append(metrics)

# Convert the list of metrics to a pandas DataFrame
df = pd.DataFrame(all_metrics)
```
